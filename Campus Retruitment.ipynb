```python
dummy_variable_1 = pd.get_dummies(data=df, columns=["Gender","Grade 10 board","Grade 12 board","Grade 12 group","Undergrad Stream","Work Ex","Specialization"])
dummy_variable_1.head()
```

```python
dummy_variable_1 = dummy_variable_1.drop(["Roll no","Grade 10 mark","Grade 12 mark","Undergrad CGPA","Employability test","MBA CGPA","Salary","Status"], axis = 1) 
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)
```

```python
df.head(5)
```

```python
print (df.columns)
print (df.shape)
```

```python
x = df[['Grade 10 mark', 'Grade 12 mark', 
        'Undergrad CGPA', 'Employability test', 'MBA CGPA', 'Gender_F', 'Gender_M',
        'Grade 10 board_Central', 'Grade 10 board_Others',
        'Grade 12 board_Central', 'Grade 12 board_Others',
        'Grade 12 group_Arts', 'Grade 12 group_Commerce',
        'Grade 12 group_Science', 'Undergrad Stream_Comm&Mgmt',
        'Undergrad Stream_Others', 'Undergrad Stream_Sci&Tech', 'Work Ex_No',
        'Work Ex_Yes', 'Specialization_Mkt&Fin', 'Specialization_Mkt&HR']].values
x[0:5]
```

```python
y = df["Status"]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_trainset, x_testset, y_trainset, y_testset = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print("The shap of trainset x :", x_trainset.shape)
print("The shap of trainset y :", y_trainset.shape)
print("The shap of trainset x :", x_testset.shape)
print("The shap of trainset y :", y_testset.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

placement = DecisionTreeClassifier(criterion="entropy", max_depth = 20)
placement # it shows the default parameters
```

```python
placement.fit(x_trainset,y_trainset)
```

```python
placement_predict = placement.predict(x_testset)
```

```python
print (placement_predict [0:10])
print (y_testset [0:10])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, placement_predict))
```

```python
i=1
temp_predict = 0
for i in range(1,21):
    temp = DecisionTreeClassifier(criterion="entropy", max_depth = i)
    temp.fit(x_trainset,y_trainset)
    temp_predict = temp.predict(x_testset)
    s =metrics.accuracy_score(y_testset, temp_predict)
    print("DecisionTrees's Accuracy for %d is %f" %(i,s))
    
```
